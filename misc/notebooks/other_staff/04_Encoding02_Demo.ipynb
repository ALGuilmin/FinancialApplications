# Encoding 2 Demo

Let $\mathbf{U}_p$ the operator for loading an array called: $p(x_i)$. The operator will act following equation $(1)$:

$$\mathbf{U}_p |0\rangle_{n} = \sum_{i=0}^{2^{n}-1}\sqrt{p(x_i)}|i\rangle_{n} \tag{1}$$

We want to know how the dagger of $\mathbf{U}_p$ acts over any basis state, this is we want to know:

$$\mathbf{U}_p \dagger |i\rangle_{n}$$

If we depicted the equation $(1)$ in matrix form:

$$
\begin{pmatrix}
u_{0 \, 0} & u_{0\,1} & \cdots & u_{0\,2^n-1}\\
u_{1\,0} & u_{1\,1} & \cdots & u_{1\,2^n-1}\\
\vdots & \vdots & \ddots & \vdots \\
u_{2^n-1\,0} & u_{2^n-1\,1} & \cdots & u_{2^n-1\,2^n-1}\\
\end{pmatrix}
\times
\begin{pmatrix}
1 \\
0 \\
\vdots \\
0
\end{pmatrix}
=
\begin{pmatrix}
u_{0 \, 0} \\
u_{1\,0} \\
\vdots \\
u_{2^n-1\,0}
\end{pmatrix}
=
\begin{pmatrix}
\sqrt{p(x_0)} \\
\sqrt{p(x_1)} \\
\vdots \\
\sqrt{p(x_{2^n-1})}
\end{pmatrix}
$$

So

$$u_{i\,0} = \sqrt{p(x_i)}$$

$$
\mathbf{U}_p =
\begin{pmatrix}
\sqrt{p(x_0)} & u_{0\,1} & \cdots & u_{0\,2^n-1}\\
\sqrt{p(x_1)} & u_{1\,1} & \cdots & u_{1\,2^n-1}\\
\vdots & \vdots & \ddots & \vdots \\
\sqrt{p(x_{2^n-1})} & u_{2^n-1\,1} & \cdots & u_{2^n-1\,2^n-1}\\
\end{pmatrix}
$$


And

$$
\mathbf{U}_p \dagger = 
\begin{pmatrix}
\sqrt{p(x_0)} & \sqrt{p(x_1)}  & \cdots & \sqrt{p(x_{2^n-1})} \\
u_{0\,1} & u_{1\,1} & \cdots & u_{2^n-1\,1}\\
\vdots & \vdots & \ddots & \vdots \\
u_{0\,2^n-1} & u_{1\,2^n-1} & \cdots & u_{2^n-1\,2^n-1}\\
\end{pmatrix}
$$


How $\mathbf{U}_p \dagger$ acts on the different basis vectors:


$$
\mathbf{U}_p \dagger |0\rangle_{n} =
\begin{pmatrix}
\sqrt{p(x_0)} & \sqrt{p(x_1)}  & \cdots & \sqrt{p(x_{2^n-1})} \\
u_{0\,1} & u_{1\,1} & \cdots & u_{2^n-1\,1}\\
\vdots & \vdots & \ddots & \vdots \\
u_{0\,2^n-1} & u_{1\,2^n-1} & \cdots & u_{2^n-1\,2^n-1}\\
\end{pmatrix}
\times
\begin{pmatrix}
1 \\
0 \\
\vdots \\
0
\end{pmatrix}
=
\begin{pmatrix}
\sqrt{p(x_0)}  \\
u_{0\,1} \\
\vdots \\
u_{0\,2^n-1}
\end{pmatrix}
=
$$

$$
=\sqrt{p(x_0)} |0\rangle_{n} +  u_{0\,1} |1\rangle_{n} + \cdots + u_{0\,2^n-1} |2^n-1\rangle_{n}
$$


$$
\mathbf{U}_p \dagger |1\rangle_{n} =
\begin{pmatrix}
\sqrt{p(x_0)} & \sqrt{p(x_1)}  & \cdots & \sqrt{p(x_{2^n-1})} \\
u_{0\,1} & u_{1\,1} & \cdots & u_{2^n-1\,1}\\
\vdots & \vdots & \ddots & \vdots \\
u_{0\,2^n-1} & u_{1\,2^n-1} & \cdots & u_{2^n-1\,2^n-1}\\
\end{pmatrix}
\times
\begin{pmatrix}
0 \\
1 \\
\vdots \\
0
\end{pmatrix}
=
\begin{pmatrix}
\sqrt{p(x_1)}  \\
u_{1\,1} \\
\vdots \\
u_{1\,2^n-1} 
\end{pmatrix}
=
$$

$$
=\sqrt{p(x_1)} |0\rangle_{n} +  u_{1\,1} |1\rangle_{n} + \cdots + u_{1\,2^n-1}  |2^n-1\rangle_{n}
$$

And so on:


$$
\mathbf{U}_p \dagger |2^n-1\rangle_{n} =
\begin{pmatrix}
\sqrt{p(x_0)} & \sqrt{p(x_1)}  & \cdots & \sqrt{p(x_{2^n-1})} \\
u_{0\,1} & u_{1\,1} & \cdots & u_{2^n-1\,1}\\
\vdots & \vdots & \ddots & \vdots \\
u_{0\,2^n-1} & u_{1\,2^n-1} & \cdots & u_{2^n-1\,2^n-1}\\
\end{pmatrix}
\times
\begin{pmatrix}
0 \\
0 \\
\vdots \\
1
\end{pmatrix}
=
\begin{pmatrix}
\sqrt{p(x_{2^n-1})}  \\
u_{2^n-1\,1} \\
\vdots \\
u_{2^n-1\,2^n-1}
\end{pmatrix}
=
$$

$$
=\sqrt{p(x_{2^n-1})} |0\rangle_{n} + u_{2^n-1\,1} |1\rangle_{n} + \cdots + u_{2^n-1\,2^n-1} |2^n-1\rangle_{n}
$$

We can summarize the before equations in following formula:

$$
\mathbf{U}_p \dagger |i\rangle_{n} = \sqrt{p(x_i)} |0\rangle_{n} + \sum_{j=1}^{2^n-1}u_{i\,j}|j\rangle_{n}
$$
